## Descripción del menú CRUD.

El presente proyecto plantea la modificación de un almacén de fármacos en una farmacia pequeña. Se consideran unicamente los siguientes datos para inventario: Nombre medicamento, Numero de tabletas, Gramos, Tipo de medicamento (pildora, pastilla,inyectable), Numero.

### 1. Definir el almacen

```python
# Definir almacen

'''
Para evitar introducir los datos cada vez que se ejecute el notebook se harán
dos definiciones de la agenda: USando la función input() (meter los datos a mano)
y la otra usando una agenda ya establecida.
'''

almacen = {}

# Estructura del almacen:
# Nombre medicamento (nom), Numero de tabletas (num_tab), Gramos (gr)
# Tipo medicamento (tip_med)

#print('Bienvenido a tu almacén \n')
nom = 'Paracetamol'
num_tab = '12'
gr = '500 mg'
tip_med = 'Pastilla'

medicamento_1 = {
    'Nombre': nom, 'Numero Tabletas': num_tab, 'Gramaje': gr, 'Tipo': tip_med
}

nom = 'Naproxeno'
num_tab = '12'
gr = '250 mg'
tip_med = 'Pildora'

medicamento_2 = {
    'Nombre': nom, 'Numero Tabletas': num_tab, 'Gramaje': gr, 'Tipo': tip_med
}

nom = 'Loratadina'
num_tab = '8'
gr = '10 mg'
tip_med = 'Pastilla'

medicamento_3 = {
    'Nombre': nom, 'Numero Tabletas': num_tab, 'Gramaje': gr, 'Tipo': tip_med
}

print('\n')

almacen = {}
print('Tu lista de medicamentos es:')
almacen[medicamento_1['Nombre']] = (medicamento_1)
almacen[medicamento_2['Nombre']] = (medicamento_2)
almacen[medicamento_3['Nombre']] = (medicamento_3)
almacen
```

## 2. Creación de menú CRUD.

## 3. Ver elementos del almacén

```python
# Función para consultar todos los datos del almacén

def imprimir_almacen():
  print('El almacén se encuentra a continuación: \n')
  for i, (key, value) in enumerate(almacen.items(), start=1):
      print(f"{i}. {key}: {value}")
```

```python
imprimir_almacen()
```

## 4. Crear nuevo registro de medicamento

```python
def registro_medicamento():
  print('Bienvenido al administrador de contactos \n')
  print('Para añadir un nuevo medicamento solicitamos: Nombre del medicamento, Numero de tabletas, Gramos, Presentación\n')
  nom = input('Ingresa el nombre: ')
  num_tab = input('Ingresa la cantidad de medicamento: ')
  gr = input('Ingresa el gramaje: ')
  tip_med = input('Ingresa la presentación del medicamento (Pastillas, Capsulas, Inyección, etc.): ')

  medicamento_nuevo = {
      'Nombre': nom, 'Numero Tabletas': num_tab, 'Gramaje': gr, 'Tipo': tip_med
  }

  almacen[medicamento_nuevo['Nombre']] = (medicamento_nuevo)

  print('Medicamento agregado con éxito...')
  print('\nSe muestra el medicamento: ')

  print(almacen[medicamento_nuevo['Nombre']])
```

```python
registro_medicamento()
```

## 5. Consulta de medicamento

```python
def consultar_medicamento():
    print('Bienvenido al consultador de medicamentos \n')
    print('Para consultar un medicamento requerimos su nombre \n')
    nombre_consulta = input('Ingresa el nombre: \n')

    for nombres in almacen.keys():
      if nombres == nombre_consulta:
        print(f'\nLos datos de la agenda para {nombres} son:', almacen[nombres])
        break
    else:
      print(f'\nMedicamento no encontrado\n')
```

```python
consultar_medicamento()
```

## 6. Actualizar Medicamento

```python
# Crear función menú para actualizar contacto
def actualizar_medicamento():
  print('Bienvenido al administrador de medicamentos \n')
  print('Para consultar un medicamento requerimos su nombre \n')
  nombre_consulta = input('Ingresa el nombre: \n')

  for nombres in almacen.keys():
    if nombres == nombre_consulta:
      print(f'Los datos del almacén para {nombres} son:', almacen[nombres])

      while True:
        print('\n +++ Menú Principal de Almacén de Medicamentos +++')
        print('1. Actualizar nombre')
        print('2. Actualizar número de tabletas')
        print('3. Actualizar gramaje')
        print('4. Actualizar tipo de medicamento')
        print('5. Salir del menú')
        opcion = int(input('\nIngresa el numero de acuerdo a tu selección: \n'))

        if opcion == 1:
          nuevo_nom = input('Ingresa el nuevo nombre: ')
          almacen[nombres]['Nombre'] = nuevo_nom
          print(f'\nSe actualizó el nombre de contacto de {nombres} a {nuevo_nom}')
          print(f'Los datos de la agenda para {nombres} son:', almacen[nombres])

        elif opcion == 2:
          nuevo_num_tab = int(input('Ingresa el nuevo número de tabletas: '))
          almacen[nombres]['Numero Tabletas'] = nuevo_num_tab
          print(f'\nSe actualizó el número de tabletas a {nuevo_num_tab}')
          print(f'Los datos de la agenda para {nombres} son:', almacen[nombres])

        elif opcion == 3:
          nuevo_gramaje = input('Ingresa el nuevo gramaje: ')
          almacen[nombres]['Gramaje'] = nuevo_gramaje
          print(f'\nSe actualizó el gramaje a {nuevo_gramaje}')
          print(f'\nLos datos de la agenda para {nombres} son:', almacen[nombres])

        elif opcion == 4:
          nuevo_tipo = input('Ingresa el nuevo tipo de medicamento: ')
          almacen[nombres]['Tipo'] = nuevo_tipo
          print(f'\nSe actualizó el gramaje a {nuevo_tipo}')
          print(f'\nLos datos de la agenda para {nombres} son:', almacen[nombres])

        elif opcion == 5:
          print(f'Nos vemos, buen día')
          break
        else:
          print(f'\nContacto no encontrado')
          print(f'Ingresa una opción válida')

  else:
    print(f'\nContacto no encontrado')
```

```python
actualizar_medicamento()
```

## 7. Eliminar Medicamento

```python
# Crear función eliminar medicamento

def eliminar_medicamento():
  print('Bienvenido al eliminador de contactos \n')
  print('A continuación se muestran los contactos de la agenda\n')
  imprimir_almacen()
  print('\n')
  print('\nPara eliminar un medicamento requerimos su nombre \n')
  nombre_consulta = input('Ingresa el nombre: \n')

  for nombres in almacen.keys():
    if nombres == nombre_consulta:
      print(f'\nSe eliminará el medicamento {nombres}...')
      almacen.pop(nombres)
      break

  else:
    print(f'\nMedicamento no encontrado')

  #print(f'\nLos datos actualizados del almacén son: \n \t')
  imprimir_almacen()
```

```python
eliminar_medicamento()
```

## 8. Implementar funciones en menú CRUD

```python
# Crear función menú opciones
def menu():

  while True:
    print('\nBienvenido al menú CRUD de farmacia\n')
    print('\n +++ Menú Principal de Farmacia +++')
    print('1. Ver medicamentos en almacén')
    print('2. Crear nuevo registro de medicamento')
    print('3. Consultar medicamento')
    print('4. Actualizar medicamento')
    print('5. Eliminar medicamento')
    print('6. Salir del menú')
    opcion = int(input('\nIngresa el numero de acuerdo a tu selección: \n'))

    # Ver medicamentos
    if opcion == 1:
      imprimir_almacen()

    # Crear nuevo registro de medicamentos
    elif opcion == 2:
      registro_medicamento()

    # Consultar medicamento
    elif opcion == 3:
      consultar_medicamento()

    # Actualizar medicamento
    elif opcion == 4:
      actualizar_medicamento()

    # Eliminar medicamento
    elif opcion == 5:
      eliminar_medicamento()

  # Salir Menú
    elif opcion == 6:
      print(f'Nos vemos, buen día')
      break
    else:
      print(f'\nMedicamento no encontrado\n')
      print(f'Ingresa una opción válida')
```

```python
menu()
```

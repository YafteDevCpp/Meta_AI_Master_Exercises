## **SELECCIÓN**

Una **estructura de selección** es una **declaración condicional**, el cual permite que diferentes secuencias de código se ejecuten dependiendo de si una **condición booleana** especificada se evalúa como **verdadera o falsa.**

La sintaxis básica de una condicional es la siguiente:

```
if condicion1:
    # código a ejecutar si la condicion1 es verdadera
elif condicion2: (opcional)
    # código a ejecutar si la condicion1 es falsa pero la condicion2 es verdadera
else:
    # código a ejecutar si tanto la condicion1 como la condicion2 son falsas
```

### **EJEMPLO 1:**

Determinar calificación de un examen:

````python
# Definir variable
calif = 5.9
````

````python
# Definir condicional
if(calif < 6):
  print("Reprobado")
else:
  print("Aprobado")
````

### **EJEMPLO 2:**

Insertar números pares dentro de una lista:

````python
# Definir lista
pares = []
nums = list(range(1,101)) # cast
print(nums)
````

````python
# Definir condicional
for num in nums:
  if(num % 2 == 0):
    pares.append(num)
  print("Seguimos en el for")
print(pares)
````

## **ITERACIÓN**

En Python, una **estructura de iteración** es un **ciclo** que generalmente se refiere a dos tipos de **estructuras de control de flujo**: los bucles **for** y **while**.


*   **Bucle For**: El bucle for se usa para **recorrer sobre una secuencia** (que puede ser una lista, una tupla, un diccionario, un conjunto o una cadena) **o cualquier otro objeto iterable**. Cada elemento de la secuencia se procesa en el cuerpo del bucle, uno tras otro.

*   **Bucle While**: El bucle while se usa para **ejecutar repetidamente un bloque de código** siempre que una condición especificada sea verdadera.

### **CICLO FOR**

También conocido como **bucle for**, es un tipo de estructura de control de flujo en Python que permite repetir un bloque de código un número determinado de veces.

El ciclo for tiene la siguiente estructura:

```
for variable in secuencia:
    # cuerpo del ciclo

```

#### **EJEMPLO 1:**

Sumar los números del 1 a 10:

````python
# realiza el ciclo for
one_to_ten = range(1, 11)
sum = 0
for num in one_to_ten:
  sum += num
  print(sum)
print(f"Suma final: {sum}")
````

#### **EJEMPLO 2:**

Insertar numeros secuenciales dentro de una lista:

````python
# definir variables
pares = []
nums = list(range(1,101)) # cast
print(nums)
````

````python
# realizar ciclo for
for num in nums:
  if(num % 2 == 0):
    pares.append(num)
print(pares)
````

#### **EJEMPLO 3:**

Imprimir las claves y valores de un diccionario:

````python
# define tu variable
alumnos = {
    'alumno1':{
        "nombre":"Jose",
        "calificacion":5
        },
    'alumno2':{
        "nombre":"Jonathan",
        "calificacion":9
        },
    'alumno3':{
        "nombre":"Balam",
        "calificacion":7
        }
    }
````

````python
# realiza el ciclo for
for alumno in alumnos:
  for key, val in alumnos[alumno].items():
    print(f"{key}: {val}")
````

````python
for alumno in alumnos:
  print(f"{alumnos[alumno]['nombre']}: {alumnos[alumno]['calificacion']}")
````

### **CICLO WHILE**

También conocido como **bucle while**, es un tipo de estructura de control de flujo en Python que permite repetir un bloque de código mientras se cumpla una condición específica.

El ciclo for tiene la siguiente estructura:

```
while condicion:
    # código a ejecutar mientras se cumpla la condición
```

#### **EJEMPLO 1:**

Validar una cadena:

````python
# realiza el ciclo while
song = "estas son las mañanitas"
while len(song) < 100: # 100 caracteres
  print(song)
  song += "s"

print(len(song))
````

#### **EJEMPLO 2:**

Recorrer las opciones de un menú:

````python
# Definir variables
menu = ['spagetti', 'ravioli', 'pasta alfredo', 'mole', 'fetuccini', 'macarrones']
````

````python
# Realizar ciclo while
index = 0
while index < len(menu):
  print(menu[index])
  index += 1
````

````python
index = 0
while True:
  print(menu[index])
  index += 1
  if index == len(menu):
    break
  else:
    continue
````

#### **EJEMPLO 3:**

Calcular la suma de los valores de un diccionario:

````python
# define tus variables
alumnos = {
    'alumno1':{
        "nombre":"Jose",
        "calificacion":5
        },
    'alumno2':{
        "nombre":"Jonathan",
        "calificacion":9
        },
    'alumno3':{
        "nombre":"Balam",
        "calificacion":7
        }
    }
````

````python
# realiza el ciclo while
num_alumnos = len(alumnos)
sum = 0
print(num_alumnos)

while alumnos:
  elem = alumnos.popitem()
  sum += elem[1]['calificacion']

prom = sum / num_alumnos
print(prom)
````

````python
# Eliminar contacto
print('Bienvenido al administrador de contactos \n')
print('Para eliminar un contacto requerimos su nombre \n')
nombre_consulta = input('Ingresa el nombre: \n')

for nombres in agenda.keys():
  if nombres == nombre_consulta:
    print(f'Se eliminará el contacto {nombres}')
    agenda.pop(nombres)
    break

else:
  print(f'Contacto no encontrado')

print(f'Los datos de la agenda son: \n', agenda)
````

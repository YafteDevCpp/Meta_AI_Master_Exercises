```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
```

```python
x = np.arange(-1, 5, dtype=float)
y = np.array([-4.0, 1.0, 6.0, 11.0, 16.0, 21.0],dtype=float)

print(f'x = {x}, y = {y}')
```

```python
regresion_lineal = np.polyfit(x,y,1)
# y = mx + b
# beta --> m y alpha --> b
beta, alpha = regresion_lineal
print(f'y = {round(beta,1)} x + {round(alpha,1)}')
```

```python
# GraficaciГіn

plt.plot(x,y, 'o', color='green', label = 'Datos')
plt.plot(x, beta*x + alpha, color='blue', label = 'RegresiГіn Lineal')
plt.axhline(y=0, color='red')
plt.axvline(x=0, color='red')
plt.title('RegresiГіn Lineal Simple')
plt.ylabel('y')
plt.xlabel('x')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Perceptron simple

capa = tf.keras.layers.Dense(units = 1)
modelo = tf.keras.Sequential([
    tf.keras.Input(shape=(1,)),
    capa
])

modelo.compile(optimizer='sgd', loss='mse')
```

```python
print('Comenzando entrenamiento...')
historial = modelo.fit(x, y, epochs=500, verbose=True)
print('Entrenamiento finalizado')
```

```python
valor = float(input('Ingrese un valor de x para predecir y: '))
prediccion = modelo.predict(np.array([[valor]]))
print(f'El valor de y para x = {valor} es: {prediccion}')
```

$Recordando$ $la$ $ecuaciГіn$ $para$ $la$ $conversiГіn$ $de$ $grados$ $Celcius$ $a$ $Fahrenheit:$

```python
import tensorflow as tf
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow import keras
```

```python
C = np.array([-40, -10, 0, 8, 15, 22, 38])
F = np.array([-40, 14, 32, 46, 59, 72, 100])
```

```python
# Normalizando los datos de entrada
C_normalized = (C - np.min(C)) / (np.max(C) - np.min(C))
F_normalized = (F - np.min(F)) / (np.max(F) - np.min(F))

print("Celsius normalizado:", C_normalized)
print("Fahrenheit normalizado:", F_normalized)
```

```python
F1 = (C*1.8)+32

plt.scatter(C,F, label = 'Datos', color = 'red')
plt.plot(C,F1, label = 'FunciГіn')
plt.xlabel('Celsius')
plt.ylabel('Fahrenheit')
plt.title('VisualizaciГіn de ConversiГіn de grados Celsius a Fahrenheit')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# RegresiГіn Lineal
regresion = np.polyfit(C,F,1)

beta, alpha = regresion

print(f'y = {alpha} + {beta}x')
```

```python
# GraficaciГіn
plt.scatter(C,F, label = 'Datos', color = 'red')
plt.plot(C, alpha + beta*C, label = 'RegresiГіn Lineal')
plt.xlabel('Celsius')
plt.ylabel('Fahrenheit')
plt.title('RegresiГіn Lineal')
plt.grid(True)
plt.legend()
plt.show()
```

```python
modelo = tf.keras.Sequential([
    tf.keras.Input(shape=(1,)),
    tf.keras.layers.Dense(units=1)
])

modelo.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss='mse')
```

```python
print('Comenzando entrenamiento con datos normalizados...')
historial_normalized = modelo.fit(C_normalized, F_normalized, epochs=2000, verbose=False)
print('Entrenamiento finalizado con datos normalizados...')
```

```python
# Train a separate model with unnormalized data for comparison
modelo_no_normalizado = tf.keras.Sequential([
    tf.keras.Input(shape=(1,)),
    tf.keras.layers.Dense(units=1)
])

# Compile the unnormalized model with an appropriate learning rate
modelo_no_normalizado.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.01), loss='mse')

print('Comenzando entrenamiento con datos no normalizados...')
historial_no_normalizado = modelo_no_normalizado.fit(C, F, epochs=2000, verbose=False)
print('Entrenamiento finalizado con datos no normalizados')
```

```python
plt.figure(figsize=(10, 6))
plt.plot(historial_normalized.history['loss'], label='PГ©rdida con Datos Normalizados')
plt.plot(historial_no_normalizado.history['loss'], label='PГ©rdida con Datos No Normalizados')
plt.xlabel('Epoch')
plt.ylabel('Magnitud de pГ©rdida')
plt.title('ComparaciГіn de PГ©rdida de Entrenamiento (Normalizado vs. No Normalizado)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.xlabel('Epoch')
plt.ylabel('Magnitud de pГ©rdida')
plt.title('PГ©rdida de Datos Normalizados')
plt.plot(historial_normalized.history['loss'], label='PГ©rdida con Datos Normalizados')
plt.legend()
plt.grid()
plt.show()
```

```python
valor = float(input('Ingrese un valor de C para predecir F: '))

# Para predecir, tambiГ©n necesitamos normalizar el valor de entrada y luego desnormalizar la salida
valor_normalized = (valor - np.min(C)) / (np.max(C) - np.min(C))

prediccion_normalized = modelo.predict(np.array([[valor_normalized]]))

# Desnormalizar la predicciГіn de Fahrenheit
prediccion_denormalized = prediccion_normalized * (np.max(F) - np.min(F)) + np.min(F)

print(f'La predicciГіn desnormalizada para {valor}В°C es {prediccion_denormalized}В°F')
```

## **SELECCIûN**

Una **estructura de selecciû°n** es una **declaraciû°n condicional**, el cual permite que diferentes secuencias de cû°digo se ejecuten dependiendo de si una **condiciû°n booleana** especificada se evalû¤a como **verdadera o falsa.**

La sintaxis bûÀsica de una condicional es la siguiente:

```
if condicion1:
    # cû°digo a ejecutar si la condicion1 es verdadera
elif condicion2: (opcional)
    # cû°digo a ejecutar si la condicion1 es falsa pero la condicion2 es verdadera
else:
    # cû°digo a ejecutar si tanto la condicion1 como la condicion2 son falsas
```

### **EJEMPLO 1:**

Determinar calificaciû°n de un examen:

````python
# Definir variable
calif = 5.9
````

````python
# Definir condicional
if(calif < 6):
  print("Reprobado")
else:
  print("Aprobado")
````

## **ITERACIûN**

En Python, una **estructura de iteraciû°n** es un **ciclo** que generalmente se refiere a dos tipos de **estructuras de control de flujo**: los bucles **for** y **while**.


*   **Bucle For**: El bucle for se usa para **recorrer sobre una secuencia** (que puede ser una lista, una tupla, un diccionario, un conjunto o una cadena) **o cualquier otro objeto iterable**. Cada elemento de la secuencia se procesa en el cuerpo del bucle, uno tras otro.

*   **Bucle While**: El bucle while se usa para **ejecutar repetidamente un bloque de cû°digo** siempre que una condiciû°n especificada sea verdadera.

### **CICLO FOR**

Tambiûˋn conocido como **bucle for**, es un tipo de estructura de control de flujo en Python que permite repetir un bloque de cû°digo un nû¤mero determinado de veces.

El ciclo for tiene la siguiente estructura:

```
for variable in secuencia:
    # cuerpo del ciclo

```

#### **EJEMPLO 1:**

Sumar los nû¤meros del 1 a 10:

````python
# realiza el ciclo for
one_to_ten = range(1, 11)
sum = 0
for num in one_to_ten:
  sum += num
  print(sum)
print(f"Suma final: {sum}")
````

#### **EJEMPLO 2:**

Insertar numeros secuenciales dentro de una lista:

````python
# definir variables
pares = []
nums = list(range(1,101)) # cast
print(nums)
````

````python
# realizar ciclo for
for num in nums:
  if(num % 2 == 0):
    pares.append(num)
print(pares)
````

#### **EJEMPLO 3:**

Imprimir las claves y valores de un diccionario:

````python
# define tu variable
alumnos = {
    'alumno1':{
        "nombre":"Jose",
        "calificacion":5
        },
    'alumno2':{
        "nombre":"Jonathan",
        "calificacion":9
        },
    'alumno3':{
        "nombre":"Balam",
        "calificacion":7
        }
    }
````

### **CICLO WHILE**

Tambiûˋn conocido como **bucle while**, es un tipo de estructura de control de flujo en Python que permite repetir un bloque de cû°digo mientras se cumpla una condiciû°n especûÙfica.

El ciclo for tiene la siguiente estructura:

```
while condicion:
    # cû°digo a ejecutar mientras se cumpla la condiciû°n
```

#### **EJEMPLO 1:**

Validar una cadena:

````python
# realiza el ciclo while
song = "estas son las maûÝanitas"
while len(song) < 100: # 100 caracteres
  print(song)
  song += "s"

print(len(song))
````

#### **EJEMPLO 2:**

#### **EJEMPLO 3:**

Calcular la suma de los valores de un diccionario:

````python
# define tus variables
alumnos = {
    'alumno1':{
        "nombre":"Jose",
        "calificacion":5
        },
    'alumno2':{
        "nombre":"Jonathan",
        "calificacion":9
        },
    'alumno3':{
        "nombre":"Balam",
        "calificacion":7
        }
    }
````

````python
# realiza el ciclo while
num_alumnos = len(alumnos)
sum = 0
print(num_alumnos)

while alumnos:
  elem = alumnos.popitem()
  sum += elem[1]['calificacion']

prom = sum / num_alumnos
print(prom)
````
